```python
# Calculate key metrics

# Average purchase value
avg_purchase_value = df['MntTotal'].mean()

# Total frequency of purchases (web + catalog + store)
df['PurchaseFrequency'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']
avg_frequency = df['PurchaseFrequency'].mean()

# Prepare summary metrics
summary = {
    'Average Purchase Value': avg_purchase_value,
    'Average Frequency of Purchases': avg_frequency
}

summary
```

```python
# find total by visits month
plt.figure(figsize=(10, 6))
plt.plot(data.index.astype(str), data.values, marker='o')
plt.title("Mnt Total by Num Web Visits Month")
plt.xlabel("NumWebVisitsMonth")
plt.ylabel("MntTotal")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

# conclusion: 2nd month have highest Mnt Total which is more than 2 lakh

```python
#Customer Segmentation: Utilize clustering algorithms (e.g., K-means) to segment customers based on behavior and purchase patterns.
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Select relevant features for clustering
features = df[['MntTotal', 'PurchaseFrequency', 'Income', 'NumWebVisitsMonth']]

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Determine the optimal number of clusters using the Elbow method
inertia = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(features_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

# conclusion: The Elbow Curve is displayed above, showing the inertia for different numbers of clusters. The "elbow point" (where the rate of decrease in inertia slows) suggests the optimal number of clusters.

```python
def get_marital_status(row):
    if row['marital_Divorced'] == 1:
        return 'Divorced'
    elif row['marital_Married'] == 1:
        return 'Married'
    elif row['marital_Single'] == 1:
        return 'Single'
    elif row['marital_Together'] == 1:
        return 'Together'
    elif row['marital_Widow'] == 1:
        return 'Widow'
    else:
        return 'Unknown'
df['Marital'] = df.apply(get_marital_status, axis=1)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Marital', y='MntTotal', data=df, palette='viridis')
plt.title('MntTotal by marital status')
plt.xlabel('Marital status')
plt.ylabel('MntTotal')
```
