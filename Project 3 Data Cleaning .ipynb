```python
from scipy import stats

# Calculate Z-scores
z_scores = stats.zscore(df['comment_count'])

# Identify outliers (Z-score > 3 or < -3)
outliers = df[abs(z_scores) > 3]
outliers
```

```python
# Calculate IQR
Q1 = df['comment_count'].quantile(0.25)
Q3 = df['comment_count'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df[(df['comment_count'] < (Q1 - 1.5 * IQR)) | (df['comment_count'] > (Q3 + 1.5 * IQR))]
outliers
```

```python
# Cap values at the 5th and 95th percentile
#Sometimes, instead of removing outliers, you can cap them at a maximum or minimum value.
lower_bound = df['comment_count'].quantile(0.05)
upper_bound = df['comment_count'].quantile(0.95)
df['comment_count'] = df['comment_count'].clip(lower=lower_bound, upper=upper_bound)
```

```python
sns.boxenplot(y='comment_count',data=df)
```

```python
#3. Transforming the Data
#Applying a logarithmic or square root transformation can help reduce the impact of outliers, especially for data with a skewed distribution.

# Apply log transformation
df['views'] = np.log1p(df['views'])  # log(1 + value) to avoid log(0)

# Alternatively, apply square root transformation
df['views'] = np.sqrt(df['views'])
```
