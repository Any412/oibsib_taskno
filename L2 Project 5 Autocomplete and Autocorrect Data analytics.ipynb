```python
import re
from collections import defaultdict, Counter

class Autocomplete:
    def __init__(self, n=3):
        self.n = n  # n-gram size
        self.ngrams = defaultdict(Counter)

    def tokenize(self, text):
        # Tokenize text into words, removing punctuation and lowering case
        return re.findall(r'\b\w+\b', text.lower())

    def train(self, corpus):
        # Train the model with a given text corpus
        tokens = self.tokenize(corpus)
        for i in range(len(tokens) - self.n + 1):
            context = tuple(tokens[i:i + self.n - 1])
            next_word = tokens[i + self.n - 1]
            self.ngrams[context][next_word] += 1

    def predict(self, context, top_k=5):
        # Predict the next word given the context
        context = tuple(self.tokenize(context)[-self.n + 1:])
        predictions = self.ngrams.get(context, {})
        return [word for word, _ in predictions.most_common(top_k)]

if __name__ == "__main__":
    # Example usage
    corpus = "This is an example. This is another example. Autocomplete is fun."
    autocomplete = Autocomplete(n=3)
    autocomplete.train(corpus)

    context = "This is"
    predictions = autocomplete.predict(context)
    print(f"Predictions for '{context}': {predictions}")
```

```python
# Class distribution: Fraud (1) vs Non-Fraud (0)
class_counts = df['Class'].value_counts()

# Plotting class distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=class_counts.index, y=class_counts.values, palette="viridis")
plt.title("Class Distribution (Fraud vs Non-Fraud)", fontsize=14)
plt.xlabel("Class (0: Non-Fraud, 1: Fraud)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(True,alpha=0.5)
plt.xticks([0, 1], ['Non-Fraud', 'Fraud'])
plt.show()

# Percentage of fraudulent transactions
fraud_percentage = (class_counts[1] / class_counts.sum()) * 100
fraud_percentage
```
